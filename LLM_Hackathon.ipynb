## Extracting text from PDF

```python
import os
import json
import fitz  # PyMuPDF
from transformers import pipeline
from typing import List, Dict

# ===================== Step 1: Slide Content Extraction from PDF =====================
def extract_text_from_pdf(pdf_path: str) -> Dict:
    doc = fitz.open(pdf_path)
    slides_data = []
    for i, page in enumerate(doc):
        text = page.get_text().strip()
        slides_data.append({
            "slide_number": i + 1,
            "content": text
        })
    return {"file": os.path.basename(pdf_path), "slides": slides_data}
```

```python
# ===================== Step 2: Format Slide Content =====================
def format_slide_content(slide_data: Dict) -> str:
    formatted = []
    for slide in slide_data["slides"]:
        formatted.append(f"Slide {slide['slide_number']}\n")
        formatted.append(slide["content"])
        formatted.append("\n")
    return "\n".join(formatted)
```

```python
# ===================== Step 3: Combine with Transcription =====================
def load_transcription(transcription_path: str) -> str:
    with open(transcription_path, 'r') as f:
        return f.read()

def combine_transcript_and_slides(transcript: str, slide_text: str) -> str:
    combined = (
        "Lecture Transcript:\n" + transcript +
        "\n\nSlide Summary:\n" + slide_text
    )
    return combined
```

```python
# ===================== Step 4: Note Generation using LLM =====================
summarizer = pipeline("summarization", model="philschmid/bart-large-cnn-samsum")

def generate_notes(combined_text: str, max_chunk_length: int = 1000) -> str:
    paragraphs = combined_text.split("\n\n")
    chunks = []
    current_chunk = ""
    for para in paragraphs:
        if len(current_chunk) + len(para) < max_chunk_length:
            current_chunk += para + "\n\n"
        else:
            chunks.append(current_chunk.strip())
            current_chunk = para + "\n\n"
    if current_chunk:
        chunks.append(current_chunk.strip())

    notes = []
    for chunk in chunks:
        summary = summarizer(chunk, max_length=200, min_length=50, do_sample=False)[0]['summary_text']
        notes.append(summary)
    return "\n\n".join(notes)
```

```python
# ===================== Step 5: Full Pipeline Function =====================
def generate_lecture_notes(pdf_path: str, transcription_path: str, output_path: str):
    print("Extracting slide content from PDF...")
    slide_data = extract_text_from_pdf(pdf_path)

    print("Formatting slide content...")
    formatted_slides = format_slide_content(slide_data)

    print("Loading transcription...")
    transcript = load_transcription(transcription_path)

    print("Combining transcript and slides...")
    combined = combine_transcript_and_slides(transcript, formatted_slides)

    print("Generating notes using LLM...")
    final_notes = generate_notes(combined)

    with open(output_path, 'w') as f:
        f.write(final_notes)
    print(f"Lecture notes saved to {output_path}")
```

## PPTx converted to PDFs

```python
import os

def pptx_to_pdf(pptx_file: str, output_pdf: str):
    # Convert pptx to pdf using unoconv
    os.system(f"unoconv -f pdf {pptx_file}")
    print(f"PDF saved as {output_pdf}")

# Example usage:
pptx_file = "/content/Class6_Unit3_Trees_ThreadBST.pptx"
output_pdf = "/content/Class6_Unit3_Trees_ThreadBST.pdf"

pptx_to_pdf(pptx_file, output_pdf)
```
