### 🧠 JSON Extraction and Cypher Generation for Neo4j

This block defines functions to extract JSON data from text and convert structured character data into Cypher commands for populating a Neo4j graph.

---

#### **1. `extract_json(text)`**
- **Purpose:** Extracts the first JSON object or array from a text string.
- **Approach:**
  - Uses regex to find JSON structures (`{}` or `[]`).
  - Attempts to parse with `json.loads`.
  - If parsing fails, attempts a fix by removing trailing commas before re-parsing.
- **Raises:** `ValueError` if no JSON is found.

---

#### **2. `run_cypher_commands(driver, commands)`**
- **Purpose:** Executes multiple Cypher commands using a Neo4j driver.
- **Details:**
  - Opens a session.
  - Runs commands in a transaction using `execute_write`.
  - Ensures atomic execution of all queries.

---

#### **3. `character_json_to_cypher(character_data)`**
- **Purpose:** Converts character metadata into a series of Cypher statements.
- **Key Actions:**
  - **Node Creation:** Handles `Character`, `Trait`, `Location`, `Item`, and `Goal` nodes.
  - **Relationship Creation:** Establishes relationships like:
    - `HAS_TRAIT`
    - `LIVES_IN`
    - `HOLDS`
    - `HAS_GOAL`
    - Custom relationships between characters (e.g., `FRIEND`, `ENEMY`)
  - **Deduplication:** Tracks created entities to avoid duplicates.
  - **Sanitization:** Cleans strings to create valid Cypher identifiers and properties.
- **Output:** Returns the generated Cypher statements as a string. Also prints each statement for debugging.

---


````python
def extract_json(text):
    """
    Finds the first JSON object or array in `text` and returns it as Python data.
    Raises ValueError if no valid JSON is found.
    """
    # Regex to locate a top‑level JSON array or object
    pattern = re.compile(r'(\[.*\]|\{.*\})', re.DOTALL)
    match = pattern.search(text)
    if not match:
        raise ValueError("No JSON object or array found in text")

    json_str = match.group(1)
    try:
        return json.loads(json_str)
    except json.JSONDecodeError as e:
        # If simple load fails, try to be more lenient by stripping trailing commas
        cleaned = re.sub(r',\s*([\]\}])', r'\1', json_str)
        return json.loads(cleaned)

def run_cypher_commands(driver, commands):
    with driver.session() as session:
        def execute_transaction(tx):
            for command in commands:
                tx.run(command)
        session.execute_write(execute_transaction)
    print("All commands executed.")

def character_json_to_cypher(character_data):
    cypher_statements = []
    created_nodes = set()
    
    def clean_identifier(text):
        """Clean a string to make it valid as a Neo4j node identifier"""
        # Replace any character that isn't alphanumeric or underscore with underscore
        import re
        return re.sub(r'[^a-zA-Z0-9_]', '_', str(text))
    
    for char in character_data:
        char_name = char['name'].replace('"', '\\"')
        role = char.get('role', 'Character').replace('"', '\\"')
        description = char.get('description', '').replace('"', '\\"')
        char_id = clean_identifier(char_name)
        
        # Create character node
        if char_name not in created_nodes:
            cypher_statements.append(
                f'CREATE (c_{char_id}:Character {{name: "{char_name}", role: "{role}", description: "{description}"}})'
            )
            created_nodes.add(char_name)
            
        # Traits
        for trait in char.get('traits', []):
            trait_clean = trait.replace('"', '\\"')
            trait_id = clean_identifier(trait_clean)
            if trait_clean not in created_nodes:
                # Debug - print the identifier to see what it looks like
                # print(f"Debug - trait_id: {trait_id}")
                cypher_statements.append(
                    f'CREATE (t_{trait_id}:Trait {{name: "{trait_clean}"}})'
                )
                created_nodes.add(trait_clean)
            cypher_statements.append(
                f'MATCH (c:Character {{name: "{char_name}"}}), (t:Trait {{name: "{trait_clean}"}}) '
                f'CREATE (c)-[:HAS_TRAIT]->(t)'
            )
            
        # Location
        if 'location' in char and char['location']:
            location = char['location'].replace('"', '\\"')
            location_id = clean_identifier(location)
            if location not in created_nodes:
                cypher_statements.append(
                    f'CREATE (l_{location_id}:Location {{name: "{location}"}})'
                )
                created_nodes.add(location)
            cypher_statements.append(
                f'MATCH (c:Character {{name: "{char_name}"}}), (l:Location {{name: "{location}"}}) '
                f'CREATE (c)-[:LIVES_IN]->(l)'
            )
            
        # Items
        for item in char.get('items', []):
            item_clean = item.replace('"', '\\"')
            item_id = clean_identifier(item_clean)
            if item_clean not in created_nodes:
                cypher_statements.append(
                    f'CREATE (i_{item_id}:Item {{name: "{item_clean}"}})'
                )
                created_nodes.add(item_clean)
            cypher_statements.append(
                f'MATCH (c:Character {{name: "{char_name}"}}), (i:Item {{name: "{item_clean}"}}) '
                f'CREATE (c)-[:HOLDS]->(i)'
            )
            
        # Goals
        for goal in char.get('goals', []):
            goal_clean = goal.replace('"', '\\"')
            goal_id = clean_identifier(goal_clean)
            if goal_clean not in created_nodes:
                cypher_statements.append(
                    f'CREATE (g_{goal_id}:Goal {{description: "{goal_clean}"}})'
                )
                created_nodes.add(goal_clean)
            cypher_statements.append(
                f'MATCH (c:Character {{name: "{char_name}"}}), (g:Goal {{description: "{goal_clean}"}}) '
                f'CREATE (c)-[:HAS_GOAL]->(g)'
            )
            
        # Relationships
        for rel in char.get('relationships', []):
            rel_type = clean_identifier(rel.get('type', '').upper())
            # Make relationship type valid by replacing spaces and special chars with underscores
            if not rel_type:
                rel_type = "RELATED_TO"  # Default relationship type
                
            target = rel.get('target', '').replace('"', '\\"')
            if not target:
                continue
                
            target_id = clean_identifier(target)
            # Ensure target node exists (as Character)
            if target not in created_nodes:
                cypher_statements.append(
                    f'CREATE (c_{target_id}:Character {{name: "{target}"}})'
                )
                created_nodes.add(target)
            cypher_statements.append(
                f'MATCH (c1:Character {{name: "{char_name}"}}), (c2:Character {{name: "{target}"}}) '
                f'CREATE (c1)-[:{rel_type}]->(c2)'
            )
    
    # Optional: Add this to debug the output
    for i, stmt in enumerate(cypher_statements):
        print(f"Statement {i}: {stmt}")
    
    return "\n".join(cypher_statements)
````

### 🧙‍♂️ `Story_Initialisation` Class — AI-Generated Narrative Setup

This class automates the creation of a fantasy story world using a language model. It generates a story introduction, extracts character, location, and goal data, and structures it for use in a narrative-driven game.

This class enables rapid creation of story-driven content, ideal for games, interactive fiction, or worldbuilding tools powered by LLMs.


````python
class Story_Initialisation:
    def __init__(self, model="meta-llama/Llama-3.2-3B-Instruct"):
        self.model = model

    def generate(self, messages):
        pipe = pipeline(
            "text-generation", 
            model = self.model,
            torch_dtype = torch.bfloat16,
            device_map = "auto",
        )
        outputs = pipe(messages, max_new_tokens=2048, temperature=1)
        return outputs[0]['generated_text'][-1]['content']
    
    def create_story(self, min_paragraphs=5, max_paragraphs=6, total_objectives=3):
        print("Creating story...")
        print(f"Minimum paragraphs:{min_paragraphs}, Maximum paragraphs: {max_paragraphs}, Objectives of story: {total_objectives}")
        
        story_prompt = f"""
Write a {min_paragraphs}-{max_paragraphs} paragraph story introduction that includes:
- At least *three named characters* who will act as key players in the story (e.g., protagonist, ally, rival, etc.)
- A *clear antagonist* whose actions create conflict and tension
- A *richly described world* featuring at least *five distinct locations* (such as cities, forests, ruins, towers, realms, etc.)
- At least four characters with a backstory in each region
- The characters' motivations and interpersonal dynamics
- Exactly {total_objectives} compelling objectives or quests that the players could pursue throughout the story, including:
    - One objective focused on *defeating the antagonist*
    - Others that are side quests or goals (e.g., retrieve an ancient relic, rescue an ally, solve a riddle to unlock a magical gate, restore peace between factions, etc.)

Ensure the introduction is immersive and sets the stage for an epic fantasy adventure. Emphasize mystery, stakes, and potential choices the players might face.
"""
        story = self.generate(messages=[{"role": "user", "content": story_prompt}])

        print('\n')
        print(story)
        print('\n')

        return story, story_prompt

    def extract_character_info(self, story, story_prompt):
        print("Extracting character information...")
        
        character_prompt = """Based on the story, create detailed descriptions for: 
        1. Protagonist (main character)
        2. Antagonist (main opponent) 
        3. Any other important characters 
        Format each description as a brief visual summary (e.g., "armored knight with red cape")"""
        
        character_discriptions = self.generate(messages=[
            {"role": "user", "content": story_prompt},
            {"role": "assistant", "content": story},
            {"role": "user", "content": character_prompt},
            ], 
        )
        
        
        print("\n")
        print(character_discriptions)
        print("\n")
        
        character_dict_prompt = """
From the story provided, extract structured character data for a knowledge graph. For each character, return:

- name: Full name or identifier of the character
- role: Narrative role (e.g., Protagonist, Antagonist, Mentor, Companion)
- description: Short visual summary
- traits: Key adjectives or personality traits
- relationships: A list of structured links to other characters, using only the specified type values
- items: List of important objects or tools they are holding or associated with
- location: Primary place or area the character lives in or originates from
- goals: List of their personal goals or motives

Only use the following *allowed relationship types* in the relationships field:
- ENEMY_OF
- FRIEND_OF
- MENTORED_BY
- LOVES
- FEARS
- TRUSTS

Only use the following *entity edges* in your graph structure:
- LIVES_IN (character -> location)
- HOLDS (character -> item)
- HAS_TRAIT (character -> trait)
- HAS_GOAL (character -> goal)

### Format:
[
  {
    "name": "Character Name",
    "role": "Protagonist",
    "description": "Visual description here",
    "traits": ["brave", "resourceful"],
    "relationships": [
        {"type": "ENEMY_OF", "target": "Malrik"},
        {"type": "TRUSTS", "target": "Elena"}
    ],
    "items": ["enchanted sword", "map of the dark woods"],
    "location": "Silverpine Village",
    "goals": ["rescue her brother", "destroy the dark crystal"]
  },
  ...
]
Return only valid JSON in this format. 
        """
        
        character_dict_discriptions =self.generate(messages=[
            {"role": "user", "content": story_prompt},
            {"role": "assistant", "content": story},
            {"role": "user", "content": character_dict_prompt},
            ])

        print('\n')
        print(character_dict_discriptions)
        print("\n")

        antagonist_name = ""
        protagonist_name = ""

        character_discriptions_json = extract_json(character_dict_discriptions)
        return character_discriptions, character_discriptions_json, character_prompt

    def extract_location_info(self, story, story_prompt, character_discriptions, character_prompt):
        print("Extracting tileset information...")
        location_prompt = "Create an exhaustive list of locations needed to create the game world."
        
        location_discriptions = self.generate(messages=[
            {"role": "user", "content": story_prompt},
            {"role": "assistant", "content": story},
            {"role": "user", "content": character_prompt},
            {"role": "assistant", "content": character_discriptions},
            {"role": "user", "content": location_prompt}
            ], 
            )
        
        print("\n")
        print(location_discriptions['choices'][0]['message']['content'])
        print("\n")

        return location_discriptions, location_prompt

    def extract_goals(self, story, story_prompt, character_discriptions, character_prompt):
        print("Extracting goals..")
        goal_prompt = """Define the game objectives: 
        1. Main goal (primary objective) 
        2. Sub-goals (steps to achieve main goal) 
        3. Reward/penalty system with points; 
        
        Format as a clear bullet list"""
        
        goal_discriptions = self.generate(model=self.model, messages=[
            {"role": "user", "content": story_prompt},
            {"role": "assistant", "content": story},
            {"role": "user", "content": character_prompt},
            {"role": "assistant", "content": character_discriptions},
            {"role": "user", "content": goal_prompt}
            ])
        
        print("\n")
        print(goal_discriptions)
        print("\n")

        return goal_discriptions, goal_prompt
````

## GraphRAG Class Overview

The `GraphRAG` class is a Retrieval-Augmented Generation (RAG) system for interactive storytelling and knowledge retrieval. It integrates with a Neo4j graph database and a quantized large language model (LLM) pipeline for efficient inference and dynamic narrative creation.

### Core Responsibilities

- **Connects to Neo4j Database:** Initializes connection to a Neo4j instance for storing and retrieving structured story data.
- **Integrates Quantized LLM Pipeline:** Loads a quantized text generation pipeline (e.g., Llama-3.2-1B-Instruct) for efficient inference.
- **Caches Entity Names:** Preloads character and location names from the graph for fast recognition.
- **Tracks Story History:** Maintains a chronological history of events for context-aware summarization.

### Key Functionalities

- **Entity Recognition and Disambiguation:**
  - Extracts characters and locations from user queries using full name matching, first name matching, and fuzzy matching.
  - Disambiguates entities with similar names using context and the LLM.

- **Knowledge Graph Queries:**
  - Provides methods to fetch character details, relationships, and story overviews from the Neo4j database.

- **Context Retrieval for RAG:**
  - Retrieves relevant context from the graph to enhance the user's query.

- **LLM-Augmented Storytelling:**
  - Generates story summaries, introductions, and events using the LLM, grounded in the knowledge graph.

- **User Action Processing:**
  - Handles user actions and calculates probabilistic outcomes, updating the story world based on character traits.

- **Database Update Execution:**
  - Executes Cypher commands to update the graph, reflecting the evolving story world.

- **Resource Management:**
  - Cleans up LLM and database resources, optimizing memory usage when closing the system.

### Typical Workflow

1. **User Query:** User asks a question or describes an action.
2. **Entity Extraction:** Identifies relevant characters/locations from the query.
3. **Context Retrieval:** Pulls context from the Neo4j database.
4. **LLM Generation:** Generates a response or story event based on the context.
5. **World Update:** Generates and executes Cypher commands to update the graph.
6. **History Update:** Logs the event for future use.

### Design Highlights

- **Memory and Performance Optimizations:** Uses quantized models and efficient GPU memory management.
- **Flexible Story Interaction:** Supports both open-ended queries and structured actions, making it ideal for interactive storytelling or games.
- **Grounded Generation:** Ensures LLM outputs are based on the knowledge graph to maintain story consistency.

**In summary:**  
The `GraphRAG` class blends the reasoning power of an LLM with the structured memory of a graph database to create dynamic, context-aware narrative experiences.


````python
class GraphRAG:
    def __init__(self, uri, username, password, model="meta-llama/Llama-3.2-1B-Instruct"):  # Changed from 3B to 1B model
        # Neo4j connection
        self.driver = GraphDatabase.driver(uri, auth=(username, password))
        self.driver.verify_connectivity()
        print("✅ Connected to Neo4j")

        # LLM setup
        self.model = model
        # Use 8-bit quantization to reduce memory footprint
        self.pipe = pipeline(
            "text-generation",
            model=model,
            torch_dtype=torch.float16,  # Use fp16 instead of bfloat16 for less memory
            device_map="auto",
            model_kwargs={"load_in_8bit": True}  # Use 8-bit quantization
        )

        # Cache entity names for quick matching
        self.character_names = self._cache_character_names()
        self.location_names = self._cache_location_names()

        # Add history tracking
        self.history = []

    def add_to_history(self, event):
        """Add an event to the story history"""
        self.history.append({
            "timestamp": time.time(),
            "event": event
        })

    def get_story_summary(self):
        """Get a summary of the story so far based on the history"""
        
        # Create a chronological account of events
        events = [item["event"] for item in self.history]
        events_text = "\n".join([f"- {event}" for event in events])
        
        # Generate a summary using the LLM
        prompt = f"""Based on these events that have occurred in the story so far:
    
    {events_text}
    
    Please provide a concise summary of the storyline, highlighting key plot developments,
    character arcs, and the current state of affairs.
    """
        
        messages = [{"role": "user", "content": prompt}]
        outputs = self.pipe(messages, max_new_tokens=1024, temperature=0.1)
        return outputs[0]["generated_text"][-1]["content"]
   
    def close(self):
        """Close the Neo4j driver connection and clean up resources"""
        self.driver.close()
        del self.pipe  # Remove reference to the pipeline
        
        # Force torch to release CUDA memory
        import gc
        gc.collect()
        torch.cuda.empty_cache()
        print("Resources released")
    
    def query_graph(self, query, params=None):
        """Execute a Cypher query and return the results"""
        with self.driver.session() as session:
            result = session.run(query, params or {})
            return [record for record in result]
    
    def _cache_location_names(self):
        """Cache all location names for faster matching"""
        query = "MATCH (l:Location) RETURN l.name AS name"
        results = self.query_graph(query)
        return [record["name"] for record in results]

    def find_mentioned_entities(self, text):
        """
        Find entities mentioned in text by matching against the knowledge graph
        Returns two lists: mentioned characters and mentioned locations
        Enhanced to handle partial names (first names) and full names
        """
        text_lower = text.lower()
        mentioned_characters = []
        mentioned_locations = []
    
        # Step 1: Direct full name matching
        for name in self.character_names:
            if name.lower() in text_lower:
                mentioned_characters.append(name)
        
        for name in self.location_names:
            if name.lower() in text_lower:
                mentioned_locations.append(name)
    
        # Step 2: First name matching for characters
        if not mentioned_characters:
            # Create a mapping of first names to full names
            first_name_map = {}
            for full_name in self.character_names:
                first_name = full_name.split()[0]  # Get first name
                if first_name not in first_name_map:
                    first_name_map[first_name] = []
                first_name_map[first_name].append(full_name)
            
            # Check for first names in text
            for first_name, full_names in first_name_map.items():
                # Look for whole word matches using word boundaries
                pattern = r'\b' + re.escape(first_name.lower()) + r'\b'
                if re.search(pattern, text_lower):
                    if len(full_names) == 1:
                        # If only one character has this first name, it's unambiguous
                        mentioned_characters.append(full_names[0])
                    else:
                        # Multiple characters share first name - add all and let context disambiguate
                        mentioned_characters.extend(full_names)
                        # Alternatively, could implement disambiguation logic here
    
        # Step 3: If still no matches, try fuzzy matching on capitalized words
        if not mentioned_characters and not mentioned_locations:
            # Extract potential entity names (capitalized words)
            potential_entities = re.findall(r'\b[A-Z][a-z]+\b', text)
            
            for entity in potential_entities:
                # Try character matching
                char_matches = get_close_matches(entity, self.character_names, n=1, cutoff=0.7)
                if char_matches:
                    mentioned_characters.append(char_matches[0])
                    continue
    
                # Try location matching
                loc_matches = get_close_matches(entity, self.location_names, n=1, cutoff=0.7)
                if loc_matches:
                    mentioned_locations.append(loc_matches[0])
                    
        # Step 4: Remove duplicates while preserving order
        mentioned_characters = list(dict.fromkeys(mentioned_characters))
        mentioned_locations = list(dict.fromkeys(mentioned_locations))
    
        return mentioned_characters, mentioned_locations

    def _cache_character_names(self):
        """
        Cache all character names and create first name index for faster matching
        """
        query = "MATCH (c:Character) RETURN c.name AS name"
        results = self.query_graph(query)
        full_names = [record["name"] for record in results]
        
        # Create first name to full name mapping for quick lookup
        self.first_name_to_full = {}
        for name in full_names:
            if name and ' ' in name:  # Only process names with spaces (first + last)
                first_name = name.split()[0]
                if first_name not in self.first_name_to_full:
                    self.first_name_to_full[first_name] = []
                self.first_name_to_full[first_name].append(name)
                
        return full_names

    def disambiguate_character(self, first_name, context=None):
        """
        Disambiguate between multiple characters with the same first name
        Using context if provided
        
        Args:
            first_name: The first name to disambiguate
            context: Optional context to help with disambiguation
            
        Returns:
            Most likely full character name or None if can't disambiguate
        """
        if first_name not in self.first_name_to_full:
            return None
            
        candidates = self.first_name_to_full[first_name]
    
        # If only one match, return it
        if len(candidates) == 1:
            return candidates[0]
            
        # If we have context, use LLM to disambiguate
        if context and len(candidates) > 1:
            # Get character details for all candidates
            candidate_details = []
            for name in candidates:
                details = self.get_character_details(name)
                if details:
                    candidate_details.append(f"{name}: {details['description'] or ''}")
            
            candidate_info = "\n".join(candidate_details)
            
            # Ask LLM to disambiguate based on context
            prompt = f"""Given this context: "{context}"
            
            And these characters with the first name "{first_name}":
            {candidate_info}
            
            Which character is most likely being referred to in the context?
            Return only the full name of the character, nothing else."""
            
            messages = [{"role": "user", "content": prompt}]
            outputs = self.pipe(messages, max_new_tokens=64, temperature=0.1)
            result = outputs[0]["generated_text"][-1]["content"].strip()
            
            # Check if the result matches one of our candidates
            for name in candidates:
                if name.lower() in result.lower():
                    return name
                    
        # Default: return the first candidate if can't disambiguate
        return candidates[0]

    def get_character_by_name(self, name, context=None):
        """
        Get character by name, handling first name or full name
        
        Args:
            name: First name or full name of character
            context: Optional context to help with disambiguation
            
        Returns:
            Character details or None if not found
        """
        # Check if it's a full name match
        if name in self.character_names:
            return self.get_character_details(name)
            
        # Check if it's a first name match
        if name in self.first_name_to_full:
            full_name = self.disambiguate_character(name, context)
            if full_name:
                return self.get_character_details(full_name)
                
        # Try fuzzy matching
        matches = get_close_matches(name, self.character_names, n=1, cutoff=0.7)
        if matches:
            return self.get_character_details(matches[0])
            
        return None
    
    def get_character_details(self, character_name):
        """Get comprehensive details about a character from the graph"""
        query = """
        MATCH (c:Character {name: $name})
        OPTIONAL MATCH (c)-[:HAS_TRAIT]->(trait:Trait)
        OPTIONAL MATCH (c)-[:LIVES_IN]->(location:Location)
        OPTIONAL MATCH (c)-[:HOLDS]->(item:Item)
        OPTIONAL MATCH (c)-[:HAS_GOAL]->(goal:Goal)
        OPTIONAL MATCH (c)-[r]->(other:Character)
        RETURN 
            c.name AS name,
            c.role AS role,
            c.description AS description,
            collect(DISTINCT trait.name) AS traits,
            collect(DISTINCT location.name) AS locations,
            collect(DISTINCT item.name) AS items,
            collect(DISTINCT goal.description) AS goals,
            collect(DISTINCT {type: type(r), target: other.name}) AS relationships
        """
        results = self.query_graph(query, {"name": character_name})
        if results:
            return results[0]
        return None
    
    def get_characters_in_location(self, location_name):
        """Get all characters in a specific location"""
        query = """
        MATCH (c:Character)-[:LIVES_IN]->(l:Location {name: $name})
        RETURN c.name AS name
        """
        results = self.query_graph(query, {"name": location_name})
        return [record["name"] for record in results]
    
    def get_story_overview(self):
        """Get a general overview of the story world"""
        # Get protagonist and antagonist
        query = """
        MATCH (c:Character)
        WHERE c.role IN ['Protagonist', 'Antagonist', 'protagonist', 'antagonist']
        RETURN c.name AS name, c.role AS role, c.description AS description
        """
        key_characters = self.query_graph(query)
        
        # Get main locations
        query = """
        MATCH (l:Location)
        RETURN l.name AS name
        LIMIT 5
        """
        main_locations = self.query_graph(query)
        
        # Get key quests/goals
        query = """
        MATCH (g:Goal)<-[:HAS_GOAL]-(c:Character)
        RETURN g.description AS description, c.name AS character
        LIMIT 5
        """
        main_goals = self.query_graph(query)
        
        return {
            "key_characters": key_characters,
            "main_locations": main_locations,
            "main_goals": main_goals
        }
    
    def retrieve_context(self, query_text):
        """
        Core RAG function: Retrieve relevant context from the graph
        based on the query text
        """
        # First, try to identify specific entities mentioned in the query
        mentioned_characters, mentioned_locations = self.find_mentioned_entities(query_text)
        
        # Initialize context list to build our retrieved information
        context = []
        
        # If we have specific entity mentions, retrieve detailed information about them
        if mentioned_characters or mentioned_locations:
            # Add character information
            for char_name in mentioned_characters:
                char_details = self.get_character_details(char_name)
                if char_details:
                    context.append(f"Character: {char_name}")
                    if char_details['role']:
                        context.append(f"Role: {char_details['role']}")
                    if char_details['description']:
                        context.append(f"Description: {char_details['description']}")
                    
                    if char_details['traits']:
                        valid_traits = [t for t in char_details['traits'] if t]
                        if valid_traits:
                            context.append(f"Traits: {', '.join(valid_traits)}")
                    
                    if char_details['locations']:
                        valid_locations = [l for l in char_details['locations'] if l]
                        if valid_locations:
                            context.append(f"Location: {', '.join(valid_locations)}")
                    
                    if char_details['items']:
                        valid_items = [i for i in char_details['items'] if i]
                        if valid_items:
                            context.append(f"Items: {', '.join(valid_items)}")
                    
                    if char_details['goals']:
                        valid_goals = [g for g in char_details['goals'] if g]
                        if valid_goals:
                            context.append(f"Goals: {', '.join(valid_goals)}")
                    
                    if char_details['relationships']:
                        rel_texts = []
                        for rel in char_details['relationships']:
                            if rel['type'] and rel['target']:
                                rel_texts.append(f"{rel['type']} {rel['target']}")
                        if rel_texts:
                            context.append(f"Relationships: {', '.join(rel_texts)}")
                    
                    context.append("")  # Empty line for readability
            
            # Add location information
            for loc_name in mentioned_locations:
                chars_in_loc = self.get_characters_in_location(loc_name)
                if chars_in_loc:
                    context.append(f"Location: {loc_name}")
                    context.append(f"Characters present: {', '.join(chars_in_loc)}")
                    context.append("")  # Empty line for readability
        
        # If no specific entities were mentioned, provide general story information
        # This aligns with the RAG principle of always retrieving something relevant
        if not context:
            # Check if the query is about the story world in general
            general_terms = ["story", "world", "plot", "who", "where", "what", "tell me about", 
                             "overview", "summary", "characters"]
            
            if any(term in query_text.lower() for term in general_terms):
                # Get overall story information
                overview = self.get_story_overview()
                
                context.append("Story Overview:")
                
                # Add protagonist/antagonist information
                for char in overview["key_characters"]:
                    context.append(f"{char['role']}: {char['name']} - {char.get('description', '')}")
                
                # Add locations
                if overview["main_locations"]:
                    context.append("\nKey Locations:")
                    for loc in overview["main_locations"]:
                        context.append(f"- {loc['name']}")
                
                # Add main goals/quests
                if overview["main_goals"]:
                    context.append("\nMain Quests/Goals:")
                    for goal in overview["main_goals"]:
                        context.append(f"- {goal['character']}: {goal['description']}")
            else:
                # For more specific queries without clear entity matches, 
                # retrieve the most central characters as context
                query = """
                MATCH (c:Character)
                OPTIONAL MATCH (c)-[r]->()
                RETURN c.name AS name, c.role AS role, count(r) AS connection_count
                ORDER BY connection_count DESC
                LIMIT 2
                """
                central_characters = self.query_graph(query)
                
                context.append("Key story characters that might be relevant to your question:")
                for char in central_characters:
                    char_details = self.get_character_details(char["name"])
                    if char_details:
                        context.append(f"Character: {char['name']} ({char_details['role']})")
                        context.append(f"Description: {char_details['description']}")
                        
                        if char_details['goals']:
                            valid_goals = [g for g in char_details['goals'] if g]
                            if valid_goals:
                                context.append(f"Goals: {', '.join(valid_goals)}")
                        
                        context.append("")  # Empty line for readability
        
        return "\n".join(context)
    
    def process_query(self, user_query):
        """
        Process a query using the RAG approach with memory optimization
        """
        # Get relevant context from the graph
        context = self.retrieve_context(user_query)

        # Create augmented prompt with retrieved context
        messages = [
            {"role": "system", "content": f"""You are a storytelling assistant with knowledge about a fictional world and its characters.
            Please answer the user's question based on the following information retrieved from the story's knowledge graph:

            {context}

            Base all of your answers on the given context only. Make only reasonable deductions from the context, do not make assumptions.
            """},
            {"role": "user", "content": user_query}
        ]

        # Use smaller max_new_tokens for memory efficiency
        try:
            outputs = self.pipe(messages, max_new_tokens=256, temperature=0.01)
            response = outputs[0]["generated_text"][-1]["content"]
        except Exception as e:
            print(f"Error during inference: {e}")
            response = "I'm sorry, I couldn't process your request due to a technical issue."
        finally:
            # Force garbage collection
            import gc
            gc.collect()
            torch.cuda.empty_cache()  # Clear CUDA cache
            
        return response
    
    def generate_story_event(self, character_names, location=None):
        """Generate a story event involving specific characters at an optional location"""
        # Build context about the characters
        context = []
        for char_name in character_names:
            char_details = self.get_character_details(char_name)
            if char_details:
                context.append(f"Character: {char_name}")
                context.append(f"Role: {char_details['role']}")
                context.append(f"Description: {char_details['description']}")
                
                if char_details['traits']:
                    valid_traits = [t for t in char_details['traits'] if t]
                    if valid_traits:
                        context.append(f"Traits: {', '.join(valid_traits)}")
                
                if char_details['items']:
                    valid_items = [i for i in char_details['items'] if i]
                    if valid_items:
                        context.append(f"Items: {', '.join(valid_items)}")
                
                if char_details['goals']:
                    valid_goals = [g for g in char_details['goals'] if g]
                    if valid_goals:
                        context.append(f"Goals: {', '.join(valid_goals)}")
                
                # Add relationships between these specific characters
                relationships = []
                for rel in char_details['relationships']:
                    if rel['target'] in character_names:
                        relationships.append(f"{char_name} {rel['type']} {rel['target']}")
                
                if relationships:
                    context.append(f"Relevant relationships: {', '.join(relationships)}")
                
                context.append("")  # Empty line for readability
        
        # Add location context if provided
        location_context = ""
        if location:
            chars_in_loc = self.get_characters_in_location(location)
            location_context = f"Location: {location}\n"
            if chars_in_loc:
                location_context += f"Characters normally present: {', '.join(chars_in_loc)}\n\n"

        context_full = '\n'.join(context)
        
        # Generate the story event
        prompt = f"""Based on the following character information:
        
        {context_full}
        
        {location_context}
        
        Generate a short but compelling story event involving {', '.join(character_names)}
        {f'at {location}' if location else ''}. The event should be consistent with their
        personalities, goals, and relationships.
        """
        
        messages = [{"role": "user", "content": prompt}]
        outputs = self.pipe(messages, max_new_tokens=1024, temperature=0.3)
        return outputs[0]["generated_text"][-1]["content"]
    
    def get_relationships_between_characters(self, character1, character2):
        """Get direct relationships between two characters"""
        query = """
        MATCH (c1:Character {name: $char1})-[r]->(c2:Character {name: $char2})
        RETURN type(r) AS relationship_type
        UNION
        MATCH (c1:Character {name: $char2})-[r]->(c2:Character {name: $char1})
        RETURN type(r) AS relationship_type
        """
        results = self.query_graph(query, {"char1": character1, "char2": character2})
        return [record["relationship_type"] for record in results]
    
    def explore_story_possibilities(self, character_name):
        """Given a character, explore possible story developments based on their goals and relationships"""
        char_details = self.get_character_details(character_name)
        if not char_details:
            return f"Character {character_name} not found in the database."
        
        # Get character's goals
        goals = char_details['goals']
        
        # Get character's relationships
        relationships = char_details['relationships']
        
        # Create context for story possibilities
        context = [
            f"Character: {character_name}",
            f"Role: {char_details['role']}",
            f"Description: {char_details['description']}"
        ]
        
        if goals:
            valid_goals = [g for g in goals if g]
            if valid_goals:
                context.append(f"Goals: {', '.join(valid_goals)}")
        
        if relationships:
            rel_texts = []
            for rel in relationships:
                if rel['type'] and rel['target']:
                    rel_texts.append(f"{rel['type']} {rel['target']}")
            if rel_texts:
                context.append(f"Relationships: {', '.join(rel_texts)}")
        
        # Generate story possibilities
        context_full = '\n'.join(context)
        prompt = f"""Based on the following character information:
        
        {context_full}
        
        Generate a possible story development or plot point that could involve this character.
        Each possibility should tie into their goals and relationships.
        """
        
        messages = [{"role": "user", "content": prompt}]
        outputs = self.pipe(messages, max_new_tokens=1024, temperature=0.3)
        return outputs[0]["generated_text"][-1]["content"]
        
    def process_user_action(self, character_name, action_description, location=None):
        """
        Process a user-described action with random success/failure mechanics.
        
        Args:
            character_name: The character performing the action
            action_description: User's description of the action they want to take
            location: Optional location where the action is taking place
            
        Returns:
            Result description and any Neo4j commands to update the graph
        """
        import random
        
        # First, determine if this is a risky action that needs success/failure evaluation
        risky_action_keywords = ["attack", "steal", "sneak", "persuade", "intimidate", "trick", 
                                "confront", "challenge", "seduce", "deceive", "break in", "escape",
                                "fight", "battle", "climb", "jump", "cast spell", "negotiate"]
        
        is_risky = any(keyword in action_description.lower() for keyword in risky_action_keywords)
        
        # Get character details for context
        char_details = self.get_character_details(character_name)
        if not char_details:
            return f"Character {character_name} not found in the database.", []
        
        # Build context for LLM to understand the character and situation
        context = [
            f"Character: {character_name}",
            f"Role: {char_details['role']}",
            f"Description: {char_details['description']}"
        ]
        
        if char_details['traits']:
            valid_traits = [t for t in char_details['traits'] if t]
            if valid_traits:
                context.append(f"Traits: {', '.join(valid_traits)}")
        
        if char_details['items']:
            valid_items = [i for i in char_details['items'] if i]
            if valid_items:
                context.append(f"Items: {', '.join(valid_items)}")
        
        # Add location context if provided
        if location:
            context.append(f"Location: {location}")
            chars_in_loc = self.get_characters_in_location(location)
            if chars_in_loc:
                context.append(f"Characters present: {', '.join(chars_in_loc)}")
        
        context_full = '\n'.join(context)
        
        # Calculate success chance based on character traits and action
        if is_risky:
            base_chance = 50  # Base 50% chance
            
            # Adjust for traits
            for trait in char_details['traits']:
                trait = trait.lower() if trait else ""
                # Positive traits boost chance
                if any(pos in trait for pos in ["strong", "clever", "skilled", "intelligent", "wise", "quick"]):
                    base_chance += 10
                # Negative traits reduce chance
                if any(neg in trait for neg in ["weak", "foolish", "clumsy", "slow", "frail"]):
                    base_chance -= 10
            
            # Ensure within bounds
            success_chance = max(min(base_chance, 95), 5)  # Between 5% and 95%
            
            # Determine outcome
            success = random.randint(1, 100) <= success_chance
            outcome_type = "success" if success else "failure"
            
            prompt = f"""Based on the following character information:
    
    {context_full}
    
    The character attempts the following action:
    "{action_description}"
    
    The attempt is a {outcome_type} (with a {success_chance}% chance of success).
    
    1. Describe in detail what happens as a result of this {outcome_type}.
    2. Determine what changes need to be made to the story world as a result, such as:
       - New relationships formed or altered
       - Items gained or lost
       - Character traits changed
       - New locations discovered or status changes to locations
       
    Output your response in 2 parts:
    1. A narrative description of what happens
    2. A JSON array of Neo4j Cypher commands to update the graph database, formatted as:
    [
      "CYPHER_COMMAND_1",
      "CYPHER_COMMAND_2",
      ...
    ]
    
    If no database changes are needed, return an empty array [].
    """
        else:
            # Non-risky actions always succeed
            prompt = f"""Based on the following character information:
    
    {context_full}
    
    The character performs the following action:
    "{action_description}"
    
    1. Describe what happens as a result of this action.
    2. Determine what changes need to be made to the story world as a result, such as:
       - New relationships formed or altered
       - Items gained or lost
       - Character traits changed
       - New locations discovered or status changes to locations
       
    Output your response in 2 parts:
    1. A narrative description of what happens
    2. A JSON array of Neo4j Cypher commands to update the graph database, formatted as:
    [
      "CYPHER_COMMAND_1",
      "CYPHER_COMMAND_2",
      ...
    ]
    
    If no database changes are needed, return an empty array [].
    """
    
        # Generate the result
        messages = [{"role": "user", "content": prompt}]
        outputs = self.pipe(messages, max_new_tokens=1024, temperature=0.3)
        response = outputs[0]["generated_text"][-1]["content"]
        
        # Extract narrative and Cypher commands
        narrative_part = response.split("```")[0] if "```" in response else response
        
        # Try to extract JSON commands - various possible formats
        cypher_commands = []
        try:
            # Look for JSON array in the response
            match = re.search(r'\[\s*".*"\s*\]', response, re.DOTALL)
            if match:
                cypher_commands = json.loads(match.group(0))
            else:
                # Try to find commands between code blocks
                code_blocks = re.findall(r'```(?:json)?\s*(.*?)\s*```', response, re.DOTALL)
                if code_blocks:
                    for block in code_blocks:
                        if block.strip().startswith("[") and block.strip().endswith("]"):
                            cypher_commands = json.loads(block)
                            break
        except (json.JSONDecodeError, ValueError) as e:
            print(f"Error parsing Cypher commands: {e}")
            # Extract commands as best-effort fallback
            cypher_lines = re.findall(r'(?:CREATE|MATCH|MERGE).*?;', response)
            cypher_commands = [line.strip() for line in cypher_lines]
        
        # Add this event to history
        if is_risky:
            outcome_word = "successfully" if success else "unsuccessfully"
            history_entry = f"{character_name} {outcome_word} attempted to {action_description}"
        else:
            history_entry = f"{character_name} performed action: {action_description}"
        
        self.add_to_history(history_entry)
        
        return narrative_part.strip(), cypher_commands
    
    def execute_action_commands(self, commands):
        """Execute the Cypher commands generated from a user action"""
        if not commands:
            return "No changes needed to the database."
        
        results = []
        with self.driver.session() as session:
            for command in commands:
                try:
                    result = session.run(command)
                    results.append(f"Executed: {command}")
                except Exception as e:
                    results.append(f"Error executing: {command}. Error: {str(e)}")
        
        return "\n".join(results)

    def get_story_intro(self):
        """Get story introduction from the RAG system with historical context"""
        # Include the original story for context
        original_story = self.history[0] if self.history else ""
        
        # Get a summary of events that have happened so far (if any)
        events_summary = ""
        if len(self.history) > 1:
            events = self.history[1:]
            events_text = "\n".join([f"- {event['event']}" for event in events])
            events_summary = f"\n\nEvents that have occurred so far:\n{events_text}"
        
        # Create a comprehensive prompt that includes original story and history
        intro_prompt = f"""Based on the following original story:

{original_story}
{events_summary}

Give me a brief, engaging introduction to this story world that:
1. Sets the stage for the adventure
2. Introduces the key conflict or tension
3. Hints at the main characters' roles
4. Creates intrigue about the world

Keep it concise but immersive - around 2-3 paragraphs that would make someone eager to explore this world.
"""
        
        return self.process_query(intro_prompt)
````

### 🎮 GraphRAG Interactive Story Game

This cell initializes and runs an interactive story-based game powered by a Neo4j graph database and a custom GraphRAG class.

#### 📝 Key Functions
- display_text(text): Shows text with a typewriter effect.
- get_available_actions(): Dynamically fetches actions based on the player’s current location, characters, and locations.
- play_game(): Main game loop handling:
  - Player movement
  - Dialogue with characters
  - Exploration and quest progression
  - Dynamic story generation using graph queries and model outputs
  - Memory optimization with gc.collect() and torch.cuda.empty_cache().

#### 🕹 Gameplay
- Player starts as a key character in a randomly chosen or protagonist-defined location.
- Options include talking, traveling, exploring, viewing quests, and performing custom actions.
- Interaction is driven by graph-based data and dynamically generated narratives.

````python
# Assuming your GraphRAG class is already defined and imported
# Create connection to Neo4j database
uri = "neo4j+s://2ac4d63b.databases.neo4j.io"
username = "neo4j"
password = "bp6OappmI43Nr9Wu6n1R3N2ECIDXP4X4PqO0l0uDlLY"

# Initialize the GraphRAG system
graph_rag = GraphRAG(uri, username, password)

def display_text(text, delay=0.03):
    """Display text with typewriter effect"""
    for char in text:
        print(char, end='', flush=True)
        time.sleep(delay)
    print()
    time.sleep(0.5)

def get_available_actions(player_location, visited_locations):
    """Get available actions based on player's current location"""
    # Query graph for characters in this location
    chars_in_location = graph_rag.get_characters_in_location(player_location)
    
    # Build available actions dictionary
    actions = {}
    
    # Add character interactions
    for i, char in enumerate(chars_in_location, 1):
        actions[f"t{i}"] = f"Talk to {char}"
    
    # Query for connected locations (this would need additional graph functionality)
    # For now, we'll use a simplified approach with a list of all locations
    all_locations = [loc["name"] for loc in graph_rag.query_graph("MATCH (l:Location) RETURN l.name AS name")]
    reachable_locations = [loc for loc in all_locations if loc != player_location]
    
    # Add movement options
    for i, loc in enumerate(reachable_locations, 1):
        actions[f"g{i}"] = f"Go to {loc}"
    
    # Add quest option
    actions["q"] = "View current quests"
    
    # Add exploration option
    actions["e"] = f"Explore {player_location}"
    
    # Add help option
    actions["h"] = "Help"
    
    # Add option to see visited locations
    if visited_locations:
        actions["v"] = "View visited locations"
    
    # Add quit option
    actions["x"] = "Exit game"

    actions["a"] = "Perform a custom action"

    actions["s"] = "Get story summary"
    
    return actions

def play_game():
    """Main game loop function with better memory management"""
    # Game state
    game_running = True
    visited_locations = set()
    active_quests = []
    inventory = []
    
    try:
        # Get story overview for initial state
        overview = graph_rag.get_story_overview()
        protagonist = None
        for char in overview["key_characters"]:
            if char["role"].lower() == "protagonist":
                protagonist = char["name"]
                break

        # If no protagonist found, pick the first character
        if not protagonist and overview["key_characters"]:
            protagonist = overview["key_characters"][0]["name"]

        # Get starting location - either from protagonist or the first location
        if protagonist:
            char_details = graph_rag.get_character_details(protagonist)
            if char_details and char_details["locations"] and char_details["locations"][0]:
                player_location = char_details["locations"][0]
            elif overview["main_locations"]:
                player_location = overview["main_locations"][0]["name"]
            else:
                player_location = "Unknown Location"
        elif overview["main_locations"]:
            player_location = overview["main_locations"][0]["name"]
        else:
            player_location = "Unknown Location"

        # Get initial quests
        if overview["main_goals"]:
            for goal in overview["main_goals"]:
                active_quests.append(goal["description"])

        # Display game intro
        clear_output(wait=True)
        print("\n" + "="*60)
        print(" "*20 + "STORY ADVENTURE")
        print("="*60 + "\n")

        # Clean up memory after getting story intro
        import gc
        gc.collect()
        torch.cuda.empty_cache()
        
        story_intro = graph_rag.get_story_intro()
        display_text(story_intro)
        print("\n" + "-"*60 + "\n")

        # Clean up after each major generation
        gc.collect()
        torch.cuda.empty_cache()

        if protagonist:
            display_text(f"You are playing as {protagonist}.")

        display_text(f"You find yourself in {player_location}.")
        visited_locations.add(player_location)

        # Main game loop
        while game_running:
            print("\n" + "-"*60)
            print(f"Current location: {player_location}")
            print("-"*60)
            
            # Get and display available actions
            actions = get_available_actions(player_location, visited_locations)
            print("\nAvailable actions:")
            for key, action in actions.items():
                print(f"[{key}] {action}")
            
            # Get player input
            choice = input("\nWhat would you like to do? ").strip().lower()
            
            # Process player choice
            if choice == "x":
                # Exit game
                display_text("Thank you for playing!")
                game_running = False
    
            elif choice == "a":
                # Custom action
                action_description = input("\nDescribe your action: ")
                if action_description:
                    result_narrative, cypher_commands = graph_rag.process_user_action(protagonist, action_description, player_location)
                    display_text(result_narrative)
            
                # Apply changes to the database
                if cypher_commands:
                    print("\nUpdating game world...")
                    result = graph_rag.execute_action_commands(cypher_commands)
                    print("Game world updated!")

            elif choice == "s":
                # Display story summary
                print("\nGenerating story summary...")
                summary = graph_rag.get_story_summary()
                display_text(summary)
                
            elif choice == "h":
                # Display help
                print("\nGame Help:")
                print("- Type the letter/number shown in brackets to perform an action")
                print("- Talk to characters to learn more about the story")
                print("- Explore locations to discover items and clues")
                print("- Complete quests to progress in the story")
                
            elif choice == "q":
                # Display quests
                print("\nActive Quests:")
                if active_quests:
                    for i, quest in enumerate(active_quests, 1):
                        print(f"{i}. {quest}")
                else:
                    print("You don't have any active quests at the moment.")
                    
            elif choice == "e":
                # Explore current location
                print("\nExploring...")
                exploration_result = graph_rag.process_query(f"What can I find when exploring {player_location}?")
                display_text(exploration_result)
                
            elif choice == "v" and visited_locations:
                # Show visited locations
                print("\nLocations you've visited:")
                for loc in visited_locations:
                    print(f"- {loc}")
                
            elif choice.startswith("t"):
                # Talk to character
                try:
                    index = int(choice[1:]) - 1
                    chars_in_location = graph_rag.get_characters_in_location(player_location)
                    if 0 <= index < len(chars_in_location):
                        character = chars_in_location[index]
                        print(f"\nTalking to {character}...")
                        
                        # Generate character dialogue
                        dialogue = graph_rag.generate_story_event([character, protagonist], player_location)
                        display_text(dialogue)
                        
                        # Check for new quests (this could be enhanced with more sophisticated logic)
                        if "quest" in dialogue.lower() or "task" in dialogue.lower() or "help" in dialogue.lower():
                            new_quest = graph_rag.explore_story_possibilities(character).split("\n")[0]
                            if new_quest and new_quest not in active_quests:
                                active_quests.append(new_quest)
                                display_text(f"\nNew quest added: {new_quest}")
                    else:
                        print("Invalid character selection.")
                except ValueError:
                    print("Invalid input. Please enter a valid option.")
                    
            elif choice.startswith("g"):
                # Go to new location
                try:
                    index = int(choice[1:]) - 1
                    all_locations = [loc["name"] for loc in graph_rag.query_graph("MATCH (l:Location) RETURN l.name AS name")]
                    reachable_locations = [loc for loc in all_locations if loc != player_location]
                    
                    if 0 <= index < len(reachable_locations):
                        new_location = reachable_locations[index]
                        print(f"\nTraveling to {new_location}...")
                        
                        # Generate travel narrative
                        travel_narrative = graph_rag.process_query(f"Describe the journey from {player_location} to {new_location}")
                        display_text(travel_narrative)
                        
                        # Update player location
                        player_location = new_location
                        visited_locations.add(player_location)
                        
                        # Describe new location
                        location_desc = graph_rag.process_query(f"Describe what {protagonist} sees upon arriving at {player_location}")
                        display_text(location_desc)
                    else:
                        print("Invalid location selection.")
                except ValueError:
                    print("Invalid input. Please enter a valid option.")
                
            else:
                print("Invalid input. Please enter a valid option.")
        
            time.sleep(1)
            
            gc.collect()
            torch.cuda.empty_cache()
            
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Make sure to clean up resources
        gc.collect()
        torch.cuda.empty_cache()
````
